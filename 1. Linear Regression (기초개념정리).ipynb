### 1. Univariable Regression

#### 1-1. Sample data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1,2,3,4])
y = np.array([2,1,4,3])
```

```python
plt.scatter(X,y)
plt.show()
```

#### 1-2. Data 변환
- scikit-learn 에서 모델 학습을 위한 데이터는 `(n,c)` 형태
- `n` : 데이터의 개수
- `c` : feature의 개수

```python
# 여기서 사용하는 데이터는 n : 4개, c : 1개 

X
```

```python
X.shape
```

```python
data = X.reshape(-1,1)
data
```

```python
data.shape
```

#### 1-3. Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

##### 1-3-1. 모델 학습하기

```python
model.fit(X=data, y=y)
# model.fit(data,y)
```

##### 1-3-2. 모델의 식 확인하기

```python
# bias 확인하기

model.intercept_
```

```python
# 회귀계수 확인하기

model.coef_
```

-> 위의 두 식을 확인하여 y = 0.9999999999999991 + 0.6 * x

##### 1-3-3. 예측하기

```python
pred = model.predict(data)
pred
```

#### 1-4. 회귀선을 plot으로 표현하기

```python
plt.scatter(X,y)
plt.plot(X, pred, color='green')
plt.show()
```

### 2. Multivariable Regression

#### 2-1. Sample data

```python
bias = 1
beta = np.array([2,3,4,5]).reshape(4,1)
noise = np.random.randn(100,1)
```

`np.random.randn(m,n)` : 평균 0, 표준편차 1의 가우시안 표준정규분포 난수 matrix array(m,n) 생성

```python
beta
```

```python
noise
```

```python
X = np.random.randn(100,4)
y = bias + X.dot(beta)
y_with_noise = y + noise
```

```python
X
```

```python
beta
```

```python
X[:10]
```

```python
y[:10]
```

#### 2-2. Multivariable Regression

```python
model = LinearRegression()
model.fit(X, y_with_noise)
```

#### 2-3. 회귀식 확인하기

```python
model.intercept_
```

```python
model.coef_
```

#### 2-4. 통계적 방법을 통한 회귀식 계산

```python
bias_X = np.array([1]*len(X)).reshape(-1,1)
stat_X = np.hstack([bias_X, X])
X_X_transpose = stat_X.transpose().dot(stat_X)
X_X_transpose_inverse = np.linalg.inv(X_X_transpose)
```

```python
stat_beta = X_X_transpose_inverse.dot(stat_X.transpose()).dot(y_with_noise)
```

```python
stat_beta
```

-> y = 1.08628922 + 1.95614344*x1 + 2.87777344*x2 + 4.11944169*x3 + 4.90713687*x4

### 3. Polynomial Regression

#### 3-1. sample data

```python
# 비선형 데이터 생성
bias = 1
beta = np.array([2,3]).reshape(2,1)
nosie = np.random.randn(100,1)
```

```python
X = np.random.randn(100,1)
X_poly = np.hstack([X, X**2])
```

```python
X_poly[:10]
```

```python
y = bias + X_poly.dot(beta)
y_with_nosie = y + noise
```
